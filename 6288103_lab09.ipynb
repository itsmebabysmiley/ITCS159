```python
import pandas
from pandas .plotting import scatter_matrix
import matplotlib .pyplot as plt
from sklearn import model_selection
from sklearn .neighbors import KNeighborsClassifier
```

```python
names=['sepal-length','sepal-width','petal-length','petal-width','class']
ds=pandas.read_csv('iris.data',names=names)
```

```python
ds.plot(kind='box',subplots=True,layout=(2,2),sharex=False,sharey=False)
plt.show()
```

```python
array = ds.values
X = array [: ,0:4]
Y = array [: ,4]
```

```python
X
```

```python
Y
```

```python
# x is sepal-length  sepal-width  petal-length  petal-width
# y is type of iris
```

```python
test_size=0.20
seed=7
X_train,X_test,Y_train,Y_test=model_selection.\
train_test_split(X,Y,\
test_size=test_size,random_state=seed)
```

```python
# train_test_split will split arrays or matrices into random train and test subsets.
```

```python
print(X_train,X_test,Y_train,Y_test)
```

```python
test_size=0.10
seed=5
X_train,X_test,Y_train,Y_test=model_selection.\
train_test_split(X,Y,\
test_size=test_size,random_state=seed)
```

```python
print(X_train,X_test,Y_train,Y_test)
```

```python
#Exercise 9
# I will choose 1 because If I remember correctly it has theory call 'the nearest neighbor'
# I think it can be either odd or even.
```

```python
irisML = KNeighborsClassifier(n_neighbors=17)
irisML.fit(X_train ,Y_train )
```

```python
prediction = irisML.predict(X_test)
```

```python
prediction
```

```python
irisML.score( X_test , Y_test )
```
