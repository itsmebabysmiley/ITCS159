```python
import numpy as np
test_image=img.copy()
lx,ly,lz=test_image.shape
X,Y=np.ogrid[0:lx,0:ly]
mask=(X-lx/2)**2+(Y-ly/2)**2>lx*ly/4;
test_image[mask]=0;
io.imshow(test_image)
io.show()
```

```python
camera_multiply = img*3
io.imshow(camera_multiply)
io.show()
```

```python
from skimage.transform import resize,rotate,swirl
resized=resize(img,(350,300),mode='constant', \
anti_aliasing=True)
io.imshow(resized)
io.show()
```

```python
resized=resize(img,(50,50),mode='constant', \
anti_aliasing=True)
io.imshow(resized)
io.show()
```

```python
rotate = rotate(img , 180)
io.imshow(rotate)
io.show()
```

```python
# Rotation angle in degrees in counter-clockwise direction.
```

```python
swirled=swirl(img,rotation=0,strength=100,radius=120,\
mode='constant')
io.imshow(swirled)
io.show()
```

```python
# Perform a swirl transformation.
# strength is a frequency of circle
# radius is how far circle will go from center
```

```python
from skimage import filters
```

```python
blurred_img=filters.gaussian(img,sigma=3,multichannel=\
True)
io.imshow(blurred_img)
io.show()
```

```python
blurred_img=filters.gaussian(img,sigma=10,multichannel=\
True)
io.imshow(blurred_img)
io.show()
```

```python
alpha= 0.15
shapened_img=img+alpha*(img-blurred_img)
io.imshow(shapened_img.astype(np.int)) #np.uint8
io.show()
io.imshow(img)
io.show()
```

```python
#enhanced image = original + amount * (original - blurred)
```
